## First euler methods program

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,t):
    return -x**3 + np.sin(t)
```

```python
a = 0.0
b = 10.0
N = 100

h = (b - a)/ N

x0 = 0.0
```

```python
t = np.arange(a,b,h)
x = np.zeros(t.size)
```

```python
for i in np.arange(t.size):
    x[i] = x[i-1]+f(x[i-1],t[i-1])*h
```

```python
plt.plot(t,x)
```

## Second Order Diff Eq

```python
def f(x,t):
    return -x**3 + np.sin(t)
```

```python
a = 0.0
b = 10.0
N = 100

h = (b - a)/ N

x0 = 0.0
v0 = 10
x[0] = x0
v[0] = v0

t = np.arange(a,b,h)
x = np.zeros(t.size)
v = np.zeros(t.size)
#print(t.shape,x.shape,v.shape)

def f1(x,v,t):
    return 9.8-0.1*v**2
```

```python
for i in np.arange(t.size-1):
    x[i+1] = x[i]+v[i]*h
    v[i+1] = v[i]+f1(x[i],v[i],t[i])*h
    
```

```python
plt.figure(1)
plt.plot(t,v)
plt.figure(2)
plt.plot(t,x)
```
