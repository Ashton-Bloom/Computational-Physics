Solve x" + x = 0 for x(0) = 1 & x'(0) = 0 using the Euler method

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0.0
b = 4*np.pi
N = 100

dt = (b - a) / N
x = np.zeros(N)
x_exact = np.zeros(N)
v = np.zeros(N)
t = np.zeros(N)
x[0] = 1
v[0] = 0
t[0] = 0
x_exact[0] = 1

for i in np.arange(N-1):
    t[i+1] = (i+1)*dt
    x[i+1] = x[i] + v[i]*dt
    v[i+1] = v[i] + (-x[i])*dt
    x_exact[i+1] = np.cos((i+1)*dt)
    
plt.plot(t,x, t,x_exact)
```

### Euler-Cromer Method

```python
a = 0.0
b = 4*np.pi
N = 1000

dt = (b - a) / N
x = np.zeros(N)
x_exact = np.zeros(N)
v = np.zeros(N)
t = np.zeros(N)
x[0] = 1
v[0] = 0
t[0] = 0
x_exact[0] = 1

for i in np.arange(N-1):
    t[i+1] = (i+1)*dt
    
    #x[i+1] = x[i] + v[i]*dt
    #v[i+1] = v[i] + (-x[i])*dt
    
    v[i+1] = v[i] + (-x[i])*dt
    x[i+1] = x[i] + v[i+1]*dt

    x_exact[i+1] = np.cos((i+1)*dt)
    
plt.plot(t,x, t,x_exact)
```

### Second Order Runge-Kutta

```python
a = 0.0
b = 10*np.pi
N = 1000

dt = (b - a) / N
x = np.zeros(N)
x_exact = np.zeros(N)
v = np.zeros(N)
t = np.zeros(N)
x[0] = 1
v[0] = 0
t[0] = 0
x_exact[0] = 1

for i in np.arange(N-1):
    t[i+1] = (i+1)*dt
    
    x_prime = x[i] + 0.5*v[i]*dt
    v_prime = v[i] + 0.5*(-x[i])*dt
    t_prime = t[i] + 0.5*dt
    
    x[i+1] = x[i] + v_prime*dt
    v[i+1] = v[i] + (-x_prime)*dt
    
    
    x_exact[i+1] = np.cos((i+1)*dt)
    
plt.plot(t,x, t,x_exact)
```
