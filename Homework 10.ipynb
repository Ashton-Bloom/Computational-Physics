## Problem 1

```python
import matplotlib.pylab as p;
from mpl_toolkits.mplot3d import Axes3D
from numpy import *;
import numpy as np

Nmax = 100 
Niter = 70 
V = zeros((Nmax, Nmax) , float)

for k in range (0, Nmax-2):
    V[k,0] = 0
    V[20,20:80] = 100
    V[80,20:80] = -100

for iter in range (Niter):
    for i in range (1,Nmax-2):
        for j in range (1,Nmax-2):
            if (i == 20 or i == 80) and (j>20 and j<80):
                continue
            V[i,j] = 0.25 *(V[i + 1, j] + V[i - 1, j] + V[i, j + 1] + V[i, j - 1])
x = range(0, Nmax - 1) 
y = range (0, Nmax - 1)
X, Y = p.meshgrid (x,y)

def functz (V): # V(x, y )
    z = V[X,Y]
    return z

Z = functz(V)
Pot_Max = np.amax(Z)
Pot_Min = np.amin(Z)
print(Pot_Max)
print(Pot_Min)


fig = p.figure() # Create figure
ax = Axes3D (fig) # Plot axes
ax.plot_wireframe (X, Y, Z, color = 'b' )
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Potential')
p.show()
```

```python
import matplotlib.pylab as p;
from mpl_toolkits.mplot3d import Axes3D
from numpy import *;
import numpy;
import numpy as np

Nmax = 100 
Niter = 70 
V = zeros((Nmax, Nmax) , float)
rho = zeros((Nmax,Nmax), float)

for k in range (0, Nmax-2):
    rho[20,20:80] = 2.4
    rho[80,20:80] = -2.4

for iter in range (Niter):
    for i in range (1,Nmax-2):
        for j in range (1,Nmax-2):
            V[i,j] = 0.25 *(V[i + 1, j] + V[i - 1, j] + V[i, j + 1] + V[i, j - 1]) + (np.pi*rho[i,j])
x = range(0, Nmax - 1) ; y = range (0,Nmax - 1)
X, Y = p.meshgrid (x,y)

def functz (V): # V(x, y )
    z = V[X,Y]
    return z

Z = functz(V)
Pot_Max = np.amax(Z)
Pot_Min = np.amin(Z)
print(np.where(Z == Pot_Max), np.where(Z == Pot_Min))
print(Z[52,20],Z[52,80])

fig = p.figure() # Create figure
ax = Axes3D (fig) # Plot axes
ax.plot_wireframe (X, Y, Z, color = 'b' ) # Red wireframe
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Potential')
p.show()
```
